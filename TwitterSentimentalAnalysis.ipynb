```python
# Sentimental Analysis
# Text Analysis: Machines go though text(natural language) and give conclusions (insights)
```

```python
import pandas
data = pandas.read_csv("https://modcom.co.ke/datasets/Tweets.csv")
data
```

```python
# Data Analysis

data.isnull().sum()
```

```python
# How many tweets we are dealing with

data.shape
```

```python
# Machine Learning
# Classification: When the Y outcome variable is a text
# Regression: When the Y outcome variable is a number

# X - Features = text
# Y - Outcome = airline sentiment (positive. negative, neutral)

X = data['text']
Y = data['airline_sentiment']
```

```python
# Split data into training(70%) and testing(30%)
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X,Y, test_size=0.3, random_state=42)
```

```python
#CountVectorizer -> Method to convert text to numerical data
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer(lowercase=True, stop_words="english")
X_train_new = vectorizer.fit_transform(X_train)
print(X_train_new)
```

```python
# Bring Machine Learning models: Classification
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
```

```python
model = DecisionTreeClassifier()
model.fit(X_train_new, Y_train)
```

```python
#CountVectorizer -> Converting X_test
X_test_new = vectorizer.transform(X_test)
tested_sentiments = model.predict(X_test_new)
print(tested_sentiments)
print(Y_test)
```

```python
# Testing the sentiment
newdata = pandas.read_csv ("https://modcom.co.ke/datasets/sample.csv")
newdata
```

```python
newdata_vectorizer = vectorizer.transform(newdata['Text'])
outcome = model.predict(newdata_vectorizer)
outcome_data = pandas.DataFrame(outcome, columns = ['Sentiment'])
outcome_data
```

```python
# Merge the 2 dataframes: outcome, new data
results = pandas.merge(newdata, outcome_data, left_index=True, right_index = True)
results
```
