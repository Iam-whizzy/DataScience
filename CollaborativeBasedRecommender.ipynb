```python
# Collaborative Recommenders
# Movie Recommenders

import pandas as pd
users = pd.read_csv("https://modcom.co.ke/datasets/file.csv")
users
```

```python
# Products/ movies/tweets
movies = pd.read_csv("https://modcom.co.ke/datasets/Movie_Id_Titles.csv")
movies
```

```python
# Merge Users with their respective movies
data = pd.merge(users, movies, on='item_id')
data
```

```python
data.describe()
```

```python
# 1. Group all the movies based on the average rating
title_mean_rating = data.groupby('title')['rating'].mean()
title_mean_rating.sort_values(ascending=False)
```

```python
# How many users rated a particular item
user_rating_count = data.groupby('title')['rating'].count()
user_rating_count.sort_values(ascending=False)
```

```python
# Display the Mean Rating and the User count
mean_rating_per_movie = pd.DataFrame(title_mean_rating)
mean_rating_per_movie['number_of_views']=pd.DataFrame(user_rating_count)
mean_rating_per_movie.sort_values('number_of_views', ascending=False)
```

```python
# Pivot Table: View Data in 3D (user_id, title, rating)

pivot = data.pivot_table(index='user_id', columns = 'title', values='rating')
pivot
```

```python
selected_movie = pivot['Godfather, The (1972)']
similar = pivot.corrwith(selected_movie)
similar_df = pd.DataFrame(similar, columns = ['Correlations'])
similar_df = similar_df.join(mean_rating_per_movie)
similar_df
```

```python
# How many people? > 200
# Correlation Above 0.5
similar_df[similar_df['number_of_views']>300].sort_values('Correlations', ascending = False).head(100)
```
