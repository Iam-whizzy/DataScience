```python
#Basket Analysis/ Association Rule Mining
# Support
# Confidence
# Lift

# Total transactions: 1000
# T. Containing Milk: 300
# T. Containing Bread: 450
# T. Containing both Milk and Bread: 150

# 1. Support of milk: T.Milk/ Total T
# 300/1000 = 0.3 and 450/1000 = 0.45

# 2. Confidence of Milk to Bread
# 150/300
# 0.5

# Confidence of Bread to Milk
# 150/450
# 0.3

# 3. Lift Milk to Bread
# Confidence Milk to Bread/ Support of Bread
# 0.5/0.45
# 1.11
```

```python
import pandas as pd
data = pd.read_csv("https://modcom.co.ke/data/datasets/store.csv")
data
```

```python
# How many countries/ regions are there?
data.groupby("Country").size()
```

```python
# Pick Transactions for the same region: France
data = data[data["Country"] == "France"]
data
```

```python
# Check for empties
data.isnull().sum()
```

```python
# Statistical Analysis
data.describe()
```

```python
data.corr()
```

```python
# Group Data - InvoiceNo, Description, Total Quantity
data.groupby(['InvoiceNo', 'Description' ])['Quantity'].sum()
```

```python
# Pivot Table: VIew the data in 3D
# 1. Index: InvoiceNO
# 2. Columns: Description
# 3. Values: Quantity

pd.set_option('display.max_columns', 300)
baskets = data.pivot_table(index='InvoiceNo', columns='Description', values = 'Quantity', fill_value = 0 )
baskets
```

```python
# Any value greater than zero means the product was bought, make it 1
def encode(x):
    if x <= 0:
       return 0
    if x>0:
       return 1
```

```python
# Applying the above fn to the data
basket_sets = baskets.applymap(encode)
basket_sets
```

```python
# Apriori ALgorithm Application
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
frequent_items_set = apriori(basket_sets, min_support=0.10, use_colnames=True)
frequent_items_set
```

```python
rules = association_rules(frequent_items_set, metric = 'lift')
rules[(rules['lift'] > 5) & (rules['confidence'] > 0.7)]
```
