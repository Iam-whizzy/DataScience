```python
#Machine Learning:
#Artificial Intelligence: A process of building applications that possess human intelligence.
#Human Intelligence: able to learn(ML), understand env(computer vision), communicate(speech rec.)...
#ML: Deals with applications learning from data.
#Deep Learning: Deals with applications learning from other sources e.g images
# Categorized into:
# 1.Supervised learning
# 2.Unsupervised learning

# Supervised Learning: It involves learning from a labeled data
# Meaning into 2 aspects: Features/ Predictors/Independent Variables/ X
#                        :Outcomes/ Predicted/ Dependent/ Y
# In ML it is categorized into: Regression(Numbers), Classification(Text/ Categorical Data)
# GOAL: Prediction

# Unsupervised Learning: Deals with extracting underlying patterns in data
# Categorized into; Clustering & Association

# Classification: Predict Patient with Diabetes(PIMA Indians)
# Problem below is supervised because the outcome variable Y Is categorical
```

```python
#Step: Read data from the source
import pandas
data = pandas.read_csv("https://modcom.co.ke/data/datasets/pima.csv")
data
```

```python
# Preprocessing and Data Analysis
# Statistical Analysis

data.describe()
```

```python
data.corr()
```

```python
# Machine Learning
# Step 1 (only supervised learning)
# Divide data into features and outcome

array = data.values
array
```

```python
# Features
X = array[:, 0:8]
X
```

```python
# Outcome
Y = array[:, 8]
Y
```

```python
# Step 2: Split Data(X,Y) into Training(70%) and Testing(30%)
from sklearn import model_selection
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
# Step 3: Import Scientific Algorithms for training the model
# Classifiers

from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.svm import SVC
from sklearn.linear_model import LogisticRegression
```

```python
#Step 4: Cross-validation
# Process to determine the best model for a particular dataset
```

```python
# Step 5: Training
model = DecisionTreeClassifier()
model.fit(X_train, Y_train)
```

```python
# Model Evaluation
# Accuracy in term of percentages
predictions = model.predict(X_test)
print(predictions)
```

```python
print(Y_test)
```

```python
# Accuracy Score: Metric for evaluating model predictions against Y_test
from sklearn.metrics import accuracy_score
print(f"The accuracy score {accuracy_score(predictions, Y_test) * 100}")
```

```python
# Prediction
features = [[17, 150, 100, 85, 101, 41, 1.4, 54]]
diagnosis = model.predict(features)
print(f"You are 70% likely to be{diagnosis}")
```
