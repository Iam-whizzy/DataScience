```python
import pandas as pd
data = pd.read_csv("https://modcom.co.ke/data/datasets/schoolcleaned.csv", parse_dates=["bday", "enrolldate", "expgradate"])
data
```

```python
#Types of inferential stats
#1. One Sample T-Test: Whether a sample represents a population
#Population: Represents the entire group (Expensive and Time Consuming)
#Sample: A portion of data that is evenly distributed to rep a population
```

```python
#Get population mean
math_students_popu_mean = data['Math'].mean()
math_students_popu_mean
```

```python
#Take the sample20 (samples)
math_sample = data['Math'].sample(20)
math_sample
```

```python
#Hypothesis Formulation
# Null HYpothesis(Ho): The sample represents the population
#ALternative HYpothesis(H1) THe sample not rep the population
#Goal: ACCEPT OR REJECT NULL HYP

from scipy.stats import ttest_1samp

statistic, pvalue = ttest_1samp(a=math_sample, popmean = math_students_popu_mean)
print(pvalue)


if pvalue > 0.05:
  print('NUll Hypothesis accepted')
else:
  print('NUll Hypothesis Rejected')
```

```python
#If alpha increases from 0.05 then we accept the null hypothesis
```

```python
#Two sample:
#The school wanted to know if there is a dignificant difference in average performance between male and female students
#Define hypothesis:
#NUll HYpothesis:  There is no significant difference
#ALternate: There is a significant difference

#Project to Try
#Environment Science(Climate Change): Test if there is significant increase in global
#temperatures over time( GLobal HIstorical  Climatology Network MOnthly) GHCN - M data

#Split data into male and female students
male_students = data[data['Gender'] == 'Male']
female_students = data[data['Gender'] == 'Female']


male_students
```

```python
from scipy.stats.distributions import alpha
#Perform to two sample on performance
from scipy.stats import ttest_ind
statistic, pvalue = ttest_ind(male_students['Reading'], female_students['Reading'])

alpha = 0.05

if pvalue > alpha:
  print('Accept the Null: THe perfomance for both genders is the same')
else:
    print('Reject the Null: THe perfomance for both genders have a signigicant difference')
```
