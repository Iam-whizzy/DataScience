```python
from gtts import gTTS
```

```python
text = "Jaywalker seems like a word that must have originated in New York City, since so many pedestrians of that metropolis seem to have no regard for crossing the street at the appointed time or place. Its place of birth is actually far from New York: the word appears to come from Kansas, or at least that is where we first see evidence of it in written use. Many of the other things we might think about jaywalker are similarly other than we might suppose."

language = 'en'
```

```python
speech = gTTS(text = text, lang = language, slow = False)
speech.save('speech.mp3')
```

```python
# Speech to text - Facebook
# Wav2Vec, Wav2Vec2Tokenizor
# Managing the audio sound
# Advanced audio processing
# Pre-Trained Model (Speech to Text)

import librosa
import torch
from transformers import Wav2Vec2ForCTC, Wav2Vec2Tokenizer
```

```python
audio, rate = librosa.load('speech.mp3', sr = 16000)
print(audio)
print(rate)
```

```python
tokenizer = Wav2Vec2Tokenizer.from_pretrained("facebook/Wav2Vec2-base-960h")
```

```python
model = Wav2Vec2ForCTC.from_pretrained("facebook/Wav2Vec2-base-960h")
```

```python
# breaking down audio to tensors

input_values = tokenizer(audio, return_tensors="pt").input_values

print(input_values)
```

```python
#probability value that we can generate from our tensor values

logits = model(input_values).logits
logits
```

```python
prediction = torch.argmax(logits, dim = -1)
prediction
```

```python
text = tokenizer.batch_decode(prediction)
text
```
