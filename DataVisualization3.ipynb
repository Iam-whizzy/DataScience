```python
#Read the Power Dataset
import pandas as pd
power_data = pd.read_csv("https://modcom.co.ke/data/datasets/power.csv", parse_dates=['Date'])
power_data
```

```python
power_data['YearOfConsumption'] = power_data['Date'].dt.year
power_data
```

```python
#Moving from descriptive analysis to

#Inferential Statistics/ Analysis (Diagnostic Analysis)
# Scipy

power_data.groupby("YearOfConsumption").size()
```

```python
#Cleaning the power data
#Removing Empties
#data.isnull().sum()
power_data['Wind'].replace({'NaN':0}, inplace=True)
power_data.groupby('Wind').size()
```
