```python
#Data Visualization
import pandas as pd
data = pd.read_csv("clean_school.csv")
data
```

```python
#Libraries
#1. Matplotlib
#2. Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.available
```

```python
#use seaborn
plt.style.use('seaborn')
#countplot, Bar, pie chart, line plot, swarm plot, heat map, displot,
```

```python
#countplot using seaborn
sns.countplot(data=data, x='Gender')#Use categorical values
plt.ylabel('Count')
plt.xlabel('Gender')
plt.title('Distribution of Gender')
```

```python
sns.countplot(data=data, x='LiveOnCampus', palette='Oranges')#Magma, Blues, Reds
plt.ylabel('Number of Students')
plt.xlabel('Students Living on Campus')
plt.title("Distribution of Student's Accomodation")
```

```python
#Barplots
import numpy as np

sns.barplot(data=data, x='Gender', y='Math', palette='magma', estimator=np.mean)
#The x axis = categorical values
# The y axis = continous values
```

```python
#distplot
sns.distplot(data['Math'], kde=True)
plt.title('Distribution of Math Perfomance')
plt.xlabel('Math Perfomance')
plt.ylabel('Frequency Distribution')
```

```python
sns.histplot(data=data, kde=True, x='Math')
plt.title('Distribution of Math Performance')
plt.xlabel('Math Performance')
plt.ylabel('Frequency Distribution')

plt.show()
```
